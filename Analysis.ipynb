### Player with most 100s, 75s, 50s and 30s
#df1.match_id.unique()

```python
runs_in_match = df1.groupby(["match_id", "batsman"]).total_runs.sum().sort_values(ascending = False).reset_index()
runs_in_match[:10]
# runs_in_match.columns
```

```python
thirtys = runs_in_match[runs_in_match.total_runs>=30].groupby('batsman').count().reset_index().drop(columns = ['total_runs'])
```

```python
thirtys.rename(columns = {'match_id' : 'count'}, inplace = True)
```

```python
thirtys.sort_values(by = ['count'], ascending = False)[:10]
```

```python
fiftys = runs_in_match[runs_in_match.total_runs>=50].groupby('batsman').count().reset_index().drop(columns = ['total_runs'])
fiftys.rename(columns = {'match_id' : 'count'}, inplace = True)
fiftys.sort_values(by = ['count'], ascending = False)[:10]
```

```python
seventys = runs_in_match[runs_in_match.total_runs>=70].groupby('batsman').count().reset_index().drop(columns = ['total_runs'])
seventys.rename(columns = {'match_id' : 'count'}, inplace = True)
seventys.sort_values(by = ['count'], ascending = False)[:10]
```

```python
hundreds = runs_in_match[runs_in_match.total_runs>=100].groupby('batsman').count().reset_index().drop(columns = ['total_runs'])
hundreds.rename(columns = {'match_id' : 'count'}, inplace = True)
hundreds.sort_values(by = ['count'], ascending = False)[:10]
```
